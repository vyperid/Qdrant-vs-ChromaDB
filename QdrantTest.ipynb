***Import Libraries***

```python
import pymupdf
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
from qdrant_client import models, QdrantClient
from sentence_transformers import SentenceTransformer
```

PDF to text and splitting into 2 sentence chunks

*a conversion is needed because in the normal PDF file there are other types of elements which are not text, also splitting the text into 2 sentences for better performance*

```python
pdf_path = '/content/deadpoetssociety.pdf'
pdf_document = pymupdf.open(pdf_path)
```

```python
def split_into_two_sentence_chunks(text):
    sentences = sent_tokenize(text)
    chunks = [' '.join(sentences[i:i+4]) for i in range(0, len(sentences), 4)]
    return chunks

all_chunks = []

for page_num in range(pdf_document.page_count):
    page = pdf_document.load_page(page_num)
    text = page.get_text()
    chunks = split_into_two_sentence_chunks(text)
    all_chunks.extend(chunks)
```

```python
for i, chunk in enumerate(all_chunks):
    print(f"Chunk {i + 1}:\n{chunk}\n")
```

```python
one_sentence_chunk = []

for page_num in range(pdf_document.page_count):
    page = pdf_document.load_page(page_num)
    text = page.get_text()
    sent = sent_tokenize(text)
    one_sentence_chunk.extend(sent)


for i, chunk in enumerate(one_sentence_chunk):
    print(f"Chunk {i + 1}:\n{chunk}\n")
```

## ***QDRANT***

```python
qdrant = QdrantClient(":memory:")
```

```python
encoder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = encoder.encode(all_chunks, show_progress_bar=True)
```

```python
embeddings
```

```python
qdrant.recreate_collection(
    collection_name = "deadpoetssociety",
    vectors_config = models.VectorParams(
        size=encoder.get_sentence_embedding_dimension(),
        distance=models.Distance.COSINE)
)
```

```python
records = [
    models.Record(
        id=i,
        vector=embeddings[i],
        payload={"text": all_chunks[i]}
    )
    for i in range(len(all_chunks))
]

qdrant.upload_records(collection_name="deadpoetssociety", records=records)
```

```python
query = "Who is going to be Neil Perry's roommate?"
query_embedding = encoder.encode([query])[0]

results = qdrant.search(
    collection_name="deadpoetssociety",
    query_vector=query_embedding,
    query_filter=None,
    limit=1
)

for result in results:
    print(result.payload["text"])
```
